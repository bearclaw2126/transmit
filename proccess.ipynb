```python
import numpy as np
from scipy.signal import spectrogram


# Replace 'your_binary_file.bin' with the path to your binary file
file_path = '/home/migue/workarea/data/UE0_12032024.dat'
# Define the data type for interleaved inphase and quadrature samples (16 bits each)
dtype = np.dtype([('inphase', np.float32), ('quadrature', np.float32)])
# Read the binary file and store its contents as a float32 array
# Read the binary file and store its contents as an array of interleaved inphase and quadrature samples
with open(file_path, 'rb') as file:
    data = np.fromfile(file, dtype=dtype)

# Convert the structured array to a complex64 array
complex_data = data['inphase'].astype(np.float32) + 1j * data['quadrature'].astype(np.float32)

# Create a spectrogram
import matplotlib.pyplot as plt

frequencies, times, Sxx = spectrogram(complex_data, fs=12.288e6,nfft=256,noverlap=128/2,window='hamming',nperseg=256,return_onesided='true')  # Adjust fs (sampling frequency) as needed
# Perform FFT
fft_size = 4096
fft_data = np.fft.fft(complex_data, n=fft_size)
fft_freqs = np.fft.fftfreq(fft_size, d=1/12.35e6)
print(f"Size of complex_data: {complex_data.size}")
# Plot FFT
plt.figure()
plt.pcolormesh(times, frequencies, 20 * np.log10(abs(Sxx)), shading='auto')
plt.colorbar(label='Intensity [dB]')
plt.title('Spectrogram UE1')
plt.xlabel('Time [s]')
plt.ylabel('Frequency [dB]')
plt.grid()
plt.show()
```
